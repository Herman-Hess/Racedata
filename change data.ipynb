```python
import csv
import re

def create_change_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(['race No', '거리', '등급', '주로', '순위', '마번', '마명', '부담중량', '기수명', '마체중', '기록', 'G-3F', 'S-1F', 'G-1F'])
        
        race_info = {}
        current_race = None
        
        for line in infile:
            if '(서울) 제' in line:
                parts = line.split()
                current_race = parts[2].replace('일', '')
                race_info['race No'] = current_race
                race_info['거리'] = parts[3].replace('M', '')
                race_info['등급'] = parts[4]
            elif '날씨:' in line and '주로:' in line:
                race_info['주로'] = line.split('주로:')[1].strip()
            elif re.match(r'^\s*\d+\s+\d+', line):
                parts = line.split()
                if len(parts) >= 15:
                    writer.writerow([
                        race_info['race No'],
                        race_info['거리'],
                        race_info['등급'],
                        race_info['주로'],
                        parts[0],  # 순위
                        parts[1],  # 마번
                        parts[2],  # 마명
                        parts[7],  # 부담중량
                        parts[8],  # 기수명
                        parts[11].split('(')[0],  # 마체중
                        parts[12],  # 기록
                        next((p for p in parts[15:] if ':' not in p), ''),  # G-3F
                        next((p for p in parts[15:] if ':' in p), ''),  # S-1F
                        parts[-1]  # G-1F
                    ])

# 파일 경로 설정
input_file = r'D:\순위예측\data\성적정보_20250126(original).txt'
output_file = r'D:\순위예측\data\경주결과변경sample3.csv'

# 함수 실행
create_change_file(input_file, output_file)
```

```python
import csv
import re

def create_change_file(input_file, output_file):
    with open(input_file, 'r', encoding='utf-8') as infile, open(output_file, 'w', newline='', encoding='utf-8') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(['race No', '거리', '등급', '주로', '순위', '마번', '마명', '부담중량', '기수명', '마체중', '기록', 'G-3F', 'S-1F', 'G-1F'])
        
        race_info = {}
        current_race = None
        
        for line in infile:
            line = line.strip()
            
            if not line:
                continue  # 빈 줄 건너뛰기
            
            print(f"Processing line: {line}")  # 디버깅 출력
            
            # 경주 정보 추출
            if '제목' in line:
                parts = line.split()
                if len(parts) < 5:
                    continue  # 데이터가 부족하면 건너뛴다
                
                current_race = parts[2].replace('년','-').replace('월','-').replace('일','').replace('(','').replace(')','')
                race_number = parts[4]
                race_number = race_number.replace('경주','')
                race_info['race No'] = f"{current_race}-{race_number}" #yyyy-mm-dd-nn"형식으로 추출
                
                print(f"Race info updated: {race_info}")  # 디버깅 출력
                
            # 경주 조건 정보 추출
            elif '(서울)' in line:
                parts = line.split()
                race_info['거리'] = parts[3].replace('M', '')
                race_info['등급'] = parts[4]           
            
            # 주로 상태 추출
            elif '날씨:' in line and '주로:' in line:
                race_info['주로'] = line.split('주로:')[1].strip()
                
                print(f"Track condition updated: {race_info['주로']}")  # 디버깅 출력
            
            # 결과 데이터 파싱
            elif re.match(r'^\d+\s+\d+', line):  # 숫자로 시작하는 행
                parts = line.split()
                
                print(f"Split parts: {parts}")  # 디버깅 출력
                
                if len(parts) < 13:  # 최소한의 데이터 개수 확인
                    print("Skipping line due to insufficient data.")
                    continue
                
                g_3f = next((p for p in parts[15:] if ':' not in p), '') if len(parts) > 15 else ''
                s_1f = next((p for p in parts[15:] if ':' in p), '') if len(parts) > 15 else ''
                g_1f = parts[-1] if len(parts) > 15 else ''
                
                writer.writerow([
                    race_info.get('race No', ''),
                    race_info.get('거리', ''),
                    race_info.get('등급', ''),
                    race_info.get('주로', ''),
                    parts[0],  # 순위
                    parts[1],  # 마번
                    parts[2],  # 마명
                    parts[6],  # 부담중량
                    parts[7],  # 기수명
                    parts[11].split('(')[0],  # 마체중
                    parts[12],  # 기록
                    g_3f,  # G-3F
                    s_1f,  # S-1F
                    g_1f   # G-1F
                ])
                print(f"Row written: {parts[0]}, {parts[1]}, {parts[2]}")  # 디버깅 출력

# 파일 경로 설정
input_file = r'D:\순위예측\data\성적정보_20250126(original).txt'
output_file = r'D:\순위예측\data\경주결과변경sample19_1.csv'

# 함수 실행
create_change_file(input_file, output_file)
```
